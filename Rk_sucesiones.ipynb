# Estructura métrica de $\mathbb{R}^k$. Sucesiones

## Producto escalar

> **Proposición 12.1**
> 
> Sean $\mathbf{u}, \mathbf{v}, \mathbf{w}$ tres vectores de $\mathbb{R}^k$, y un escalar $\lambda\in\mathbb{R}$. El producto escalar satisface las siguientes propiedades:
> 
> 1. $\mathbf{u}\cdot \mathbf{u}\geq 0$
> 2. $\mathbf{u}\cdot \mathbf{u}=0 \Longleftrightarrow \mathbf{u} = 0$
> 3. $\mathbf{u}\cdot \mathbf{v} = \mathbf{v}\cdot \mathbf{u}$
> 4. $(\lambda\mathbf{u})\cdot \mathbf{v} = \lambda(\mathbf{u}\cdot \mathbf{v})$
> 5. $(\mathbf{u} + \mathbf{v})\cdot\mathbf{w} = \mathbf{u}\cdot\mathbf{w} + \mathbf{v}\cdot\mathbf{w}$

>**Proposición 12.2**
>
>Sean $\mathbf{u}, \mathbf{v}$ dos vectores de $\mathbb{R}^k$. El coseno del ángulo $\alpha$ que forman $\mathbf{u}$ y $\mathbf{v}$ verifica
>$$\cos\alpha=\frac{\mathbf{u}\cdot \mathbf{v}}{\|\mathbf{u}\|\|\mathbf{v}\|}$$
