# Funciones de varias variables. Límites y continuidad

## Límite de una función de varias variables

La noción de límite para una función de varias variables se extiende de la definición análoga del caso escalar. Utilizando la norma euclídea como distancia.

### Definición 13.4

Sea una función 
$$
\mathbf{f} = (f_1,\dots,f_m): \mathbb{R}^k\to\mathbb{R}^m
$$
y los puntos $\mathbf{a}=(a_1,\dots,a_n)\in \mathbb{R}^k$, $\mathbf{b}=(b_1,\dots,b_m)\in \mathbb{R}^m$. Se dice que $\mathbf{b}$ es el límite de $\mathbf{f}$ cuando $\mathbf{x}$ tiende a $\mathbf{a}$ y se denota por 
$$
\lim_{\mathbf{x}\to \mathbf{a}}\mathbf{f}(\mathbf{x})=\mathbf{b}
$$
si para todo $\epsilon>0$ existe un $\delta>0$ tal que
$$
0< \lVert \mathbf{x} - \mathbf{a} \rVert < \delta \Rightarrow \lVert \mathbf{f}(\mathbf{x}) - \mathbf{f}(\mathbf{a}) \rVert<\epsilon
$$
La definición de límite anterior se puede enunciar también en términos de bolas, como se muestra a continuación.
