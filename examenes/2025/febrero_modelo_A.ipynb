>### Ejercicio 1
>
>Consideremos el siguiente sistema lineal: encontrar $(x,y,z,w)\in\mathbb{R}^4$ tal que
>$$
>\left.
>\begin{aligned}
>-5x-8y-11z-14w&=-2\\
>x+y+z+2w&=1
>\end{aligned}
>\right\rbrace
>$$
>
>Se pide lo siguiente:
>
>1. Clasificar y resolver el sistema lineal.  
>2. Determinar la matriz $A$ del sistema y calcular su matriz escalonada reducida $\operatorname{red}(A)$.  
>3. Calcular una base de $\ker(A)$.

### Solución

**1. Clasificación y resolución del sistema**

Observemos que
$$
\begin{vmatrix}
-5 & -8 \\
1 & 1
\end{vmatrix}
=(-5)\cdot 1-(-8)\cdot 1=3\neq 0
$$
Por tanto, el rango de la matriz de coeficientes coincide con el rango de la matriz ampliada y el sistema es compatible por el teorema de Rouché-Frobenius. Como el rango es $2$ y el sistema tiene $4$ incógnitas, el sistema es compatible indeterminado con $4-2=2$ parámetros libres.

Resolvemos el sistema mediante operaciones elementales sobre la matriz ampliada:
$$
\begin{aligned}
\left(
\begin{array}{cccc|c}
-5 & -8 & -11 & -14 & -2 \\
 1 &  1 &   1 &   2 &  1
\end{array}
\right)
&\approx
\left(
\begin{array}{cccc|c}
 1 &  1 &   1 &   2 &  1 \\
-5 & -8 & -11 & -14 & -2
\end{array}
\right) \\
&\approx
\left(
\begin{array}{cccc|c}
1 & 1 & 1 & 2 & 1 \\
0 & -3 & -6 & -4 & 3
\end{array}
\right)
\approx
\left(
\begin{array}{cccc|c}
1 & 1 & 1 & 2 & 1 \\
0 & 1 & 2 & \frac{4}{3} & -1
\end{array}
\right) \\
&\approx
\left(
\begin{array}{cccc|c}
1 & 0 & -1 & \frac{2}{3} & 2 \\
0 & 1 & 2 & \frac{4}{3} & -1
\end{array}
\right)
\end{aligned}
$$

Tomando $z=\lambda$ y $w=\mu$, con $\lambda,\mu\in\mathbb{R}$, se obtiene la solución general
$$
\begin{aligned}
x&=2+\lambda-\frac{2}{3}\mu \\
y&=-1-2\lambda-\frac{4}{3}\mu \\
z&=\lambda \\
w&=\mu
\end{aligned}
$$

**2. Matriz del sistema y matriz reducida**

La matriz de coeficientes del sistema es
$$
A=
\begin{pmatrix}
-5 & -8 & -11 & -14 \\
 1 &  1 &   1 &   2
\end{pmatrix}
$$
con vector de términos independientes
$$
\mathbf{b}=
\begin{pmatrix}
-2\\
1
\end{pmatrix}
$$
De la reducción anterior se obtiene
$$
\operatorname{red}(A)=
\begin{pmatrix}
1 & 0 & -1 & \frac{2}{3} \\
0 & 1 & 2 & \frac{4}{3}
\end{pmatrix}
$$

**3. Cálculo de una base de $\ker(A)$**

Para calcular $\ker(A)$ se resuelve el sistema homogéneo $A\mathbf{x}=\mathbf{0}$. A partir de $\operatorname{red}(A)$ se obtiene
$$
\begin{aligned}
x&=\lambda-\frac{2}{3}\mu \\
y&=-2\lambda-\frac{4}{3}\mu \\
z&=\lambda \\
w&=\mu
\end{aligned}
\qquad \lambda,\mu\in\mathbb{R}
$$
En forma vectorial,
$$
\begin{pmatrix}
x\\
y\\
z\\
w
\end{pmatrix}
=
\lambda
\begin{pmatrix}
1\\
-2\\
1\\
0
\end{pmatrix}
+
\mu
\begin{pmatrix}
-\frac{2}{3}\\
-\frac{4}{3}\\
0\\
1
\end{pmatrix}
$$
Por tanto, una base de $\ker(A)$ es
$$
\left\lbrace
\begin{pmatrix}
1\\
-2\\
1\\
0
\end{pmatrix}
\begin{pmatrix}
-\frac{2}{3}\\
-\frac{4}{3}\\
0\\
1
\end{pmatrix}
\right\rbrace
$$

>### Ejercicio 2
>
>Sea $F:\mathbb{R}^3\to\mathbb{R}^3$ la aplicación lineal definida por
>$$
>F(x_1,x_2,x_3)=(4x_1+x_2,\,-4x_1-x_2,\,3x_3).
>$$
>
>Se pide lo siguiente:
>
>1. Determinar los valores propios y los subespacios propios asociados.  
>2. Estudiar si la matriz asociada es diagonalizable y, en caso afirmativo, hallar la matriz diagonal $D\in\mathcal{M}_3$ y una base $A\subset\mathbb{R}^3$ respecto de la cual queda diagonalizada.

### Solución

La matriz de $F$ en la base canónica de $\mathbb{R}^3$ es
$$
A=
\begin{pmatrix}
4 & 1 & 0 \\
-4 & -1 & 0 \\
0 & 0 & 3
\end{pmatrix}.
$$

El polinomio característico es
$$
\begin{aligned}
p(\lambda)
&=\det(A-\lambda I)
=
\begin{vmatrix}
4-\lambda & 1 & 0 \\
-4 & -1-\lambda & 0 \\
0 & 0 & 3-\lambda
\end{vmatrix} \\
&=(3-\lambda)\left[(4-\lambda)(-1-\lambda)+4\right]
=(3-\lambda)\left(\lambda^2-3\lambda\right)
=-\lambda(\lambda-3)^2.
\end{aligned}
$$

Por tanto, los valores propios son
$$
\lambda_1=0,\qquad \lambda_2=3,
$$
donde $\lambda=3$ tiene multiplicidad algebraica $2$.

A continuación se determinan los subespacios propios $E_{\lambda_1}=\ker(A)$ y $E_{\lambda_2}=\ker(A-3I)$.

**1. Subespacio propio asociado a $\lambda_1=0$**

Reducimos $A$ por operaciones elementales:
$$
\begin{pmatrix}
4 & 1 & 0 \\
-4 & -1 & 0 \\
0 & 0 & 3
\end{pmatrix}
\approx
\begin{pmatrix}
4 & 1 & 0 \\
0 & 0 & 0 \\
0 & 0 & 3
\end{pmatrix}
\approx
\begin{pmatrix}
1 & \frac{1}{4} & 0 \\
0 & 0 & 1 \\
0 & 0 & 0
\end{pmatrix}.
$$
El sistema $A\mathbf{x}=\mathbf{0}$ equivale a
$$
x_1+\frac{1}{4}x_2=0,\qquad x_3=0.
$$
Tomando $x_2=\lambda$, con $\lambda\in\mathbb{R}$, se obtiene $x_1=-\frac{1}{4}\lambda$ y $x_3=0$. Por tanto,
$$
E_{\lambda_1}=\ker(A)=\left\lbrace \lambda\begin{pmatrix}-\frac{1}{4}\\[2pt]1\\[2pt]0\end{pmatrix}:\lambda\in\mathbb{R}\right\rbrace
=\left\lbrace \lambda\begin{pmatrix}1\\[2pt]-4\\[2pt]0\end{pmatrix}:\lambda\in\mathbb{R}\right\rbrace.
$$

**2. Subespacio propio asociado a $\lambda_2=3$**

Se considera
$$
A-3I=
\begin{pmatrix}
1 & 1 & 0 \\
-4 & -4 & 0 \\
0 & 0 & 0
\end{pmatrix}.
$$
Su forma escalonada reducida es
$$
\operatorname{red}(A-3I)=
\begin{pmatrix}
1 & 1 & 0 \\
0 & 0 & 0 \\
0 & 0 & 0
\end{pmatrix}.
$$
El sistema $(A-3I)\mathbf{x}=\mathbf{0}$ equivale a $x_1+x_2=0$, con $x_3$ libre. Tomando $x_1=\lambda$ y $x_3=\mu$, con $\lambda,\mu\in\mathbb{R}$, se obtiene $x_2=-\lambda$. Por tanto,
$$
E_{\lambda_2}=\ker(A-3I)
=
\left\lbrace
\lambda\begin{pmatrix}1\\[2pt]-1\\[2pt]0\end{pmatrix}
+\mu\begin{pmatrix}0\\[2pt]0\\[2pt]1\end{pmatrix}
:\lambda,\mu\in\mathbb{R}
\right\rbrace.
$$

**3. Diagonalización**

Se tiene
$$
\dim E_{\lambda_1}=1,\qquad \dim E_{\lambda_2}=2,
$$
y además $1+2=3$. Por tanto, $A$ es diagonalizable.

Tomando como base de vectores propios
$$
\mathcal{B}=
\left\lbrace
\begin{pmatrix}1\\-4\\0\end{pmatrix},
\begin{pmatrix}1\\-1\\0\end{pmatrix},
\begin{pmatrix}0\\0\\1\end{pmatrix}
\right\rbrace,
$$
la matriz de $F$ respecto de $\mathcal{B}$ es diagonal y viene dada por
$$
D=
\begin{pmatrix}
0 & 0 & 0 \\
0 & 3 & 0 \\
0 & 0 & 3
\end{pmatrix}.
$$

>### Ejercicio 3
>
>Este ejercicio consta de dos apartados. Debe elegir y responder **solamente uno de ellos**.
>
>1. Sea la función
>$$
>f(x)=\cos(2x)-2\sin x
>$$
>Se pide:
>Probar la existencia de mínimos absolutos de $f$ en el intervalo $[-\pi,\pi]$ y determinar explícitamente dichos mínimos absolutos.
>
>2. Dada la función $f:\mathbb{R}^2\to\mathbb{R}$
>$$
>f(x,y)=(-x-1)^2(y+1)^3
>$$
>Determinar su polinomio de Taylor de orden $2$ en el punto $(2,1)$.

### Solución

#### 1.

La función $f$ es continua en el intervalo compacto $[-\pi,\pi]$. Por el teorema de Weierstrass, alcanza mínimo absoluto en dicho intervalo.

Para determinar los mínimos absolutos de forma explícita, basta estudiar los puntos críticos del interior del intervalo y los extremos.

Calculamos la derivada:
$$
f'(x)=-2\sin(2x)-2\cos x
=-2\bigl(2\sin x\cos x+\cos x\bigr)
=-2\cos x\,(2\sin x+1)
$$
Por tanto, $f'(x)=0$ si y solo si
$$
\cos x=0
\quad\text{o bien}\quad
\sin x=-\frac{1}{2}
$$
En el intervalo $[-\pi,\pi]$ se tiene:
$$
\cos x=0 \iff x\in\left\lbrace -\frac{\pi}{2},\frac{\pi}{2}\right\rbrace
$$
y
$$
\sin x=-\frac{1}{2} \iff x\in\left\lbrace -\frac{\pi}{6},-\frac{5\pi}{6}\right\rbrace
$$
Por tanto, basta evaluar $f$ en los puntos
$$
\left\lbrace -\pi,\pi,-\frac{\pi}{2},\frac{\pi}{2},-\frac{\pi}{6},-\frac{5\pi}{6}\right\rbrace
$$
Se obtiene:
$$
\begin{aligned}
f(-\pi) &= \cos(-2\pi)-2\sin(-\pi)=1 \\
f(\pi)  &= \cos(2\pi)-2\sin(\pi)=1 \\
f\left(\frac{\pi}{2}\right) &= \cos(\pi)-2\sin\left(\frac{\pi}{2}\right)=-1-2=-3 \\
f\left(-\frac{\pi}{2}\right) &= \cos(-\pi)-2\sin\left(-\frac{\pi}{2}\right)=-1+2=1 \\
f\left(-\frac{\pi}{6}\right) &= \cos\left(-\frac{\pi}{3}\right)-2\sin\left(-\frac{\pi}{6}\right)=\frac{3}{2} \\
f\left(-\frac{5\pi}{6}\right) &= \cos\left(-\frac{5\pi}{3}\right)-2\sin\left(-\frac{5\pi}{6}\right)=\frac{3}{2}
\end{aligned}
$$
El mínimo absoluto de $f$ en $[-\pi,\pi]$ es $-3$, y se alcanza en $x=\dfrac{\pi}{2}$.

#### 2.

Sea $f:\mathbb{R}^2\to\mathbb{R}$ definida por
$$
f(x,y)=(-x-1)^2(y+1)^3=(x+1)^2(y+1)^3
$$
El polinomio de Taylor de orden $2$ de $f$ en el punto $\mathbf{a}=(2,1)$ viene dado por
$$
\begin{aligned}
P_2(x,y)=&\,f(2,1)+D_1f(2,1)(x-2)+D_2f(2,1)(y-1) \\
&+\frac{1}{2}\Bigl(
D_{11}f(2,1)(x-2)^2
+2D_{12}f(2,1)(x-2)(y-1)
+D_{22}f(2,1)(y-1)^2
\Bigr)
\end{aligned}
$$
Calculamos las derivadas parciales:
$$
\begin{aligned}
D_1f(x,y)&=2(x+1)(y+1)^3, & D_1f(2,1)&=48 \\
D_2f(x,y)&=3(x+1)^2(y+1)^2, & D_2f(2,1)&=108 \\
D_{11}f(x,y)&=2(y+1)^3, & D_{11}f(2,1)&=16 \\
D_{12}f(x,y)&=6(x+1)(y+1)^2, & D_{12}f(2,1)&=72 \\
D_{22}f(x,y)&=6(x+1)^2(y+1), & D_{22}f(2,1)&=108
\end{aligned}
$$
Además,
$$
f(2,1)=72
$$
Sustituyendo:
$$
\begin{aligned}
P_2(x,y)=&\,72+48(x-2)+108(y-1) \\
&+8(x-2)^2+72(x-2)(y-1)+54(y-1)^2
\end{aligned}
$$

>### Ejercicio 4.
>
>Este ejercicio consta de dos apartados. Debe elegir y responder solamente a uno de ellos:
>
>1. Calcule una primitiva de la siguiente integral indefinida
>$$
>\int e^{\sqrt{2x}}\,dx
>$$
>(Ayuda: Considere el cambio $u=\sqrt{2x}$.)
>
>2. Consideremos la integral
>$$
>\int_{T} x\,dx\,dy
>$$
>donde $T$ es el triángulo formado por los puntos $(-1,-1)$, $(0,0)$ y $(1,-1)$.
>
>En este apartado se pide lo siguiente:
>* Dé una representación de tipo II del recinto de integración.
>* Calcule la integral.


### Solución

### 1.

Buscamos una función $f:[0,\infty)\to\mathbb{R}$ derivable tal que $f'(x)=e^{\sqrt{2x}}$. Por el primer teorema fundamental del cálculo, una posible elección es
$$
f(x)=\int_{0}^{x} e^{\sqrt{2t}}\,dt
$$

Hacemos el cambio $u=\sqrt{2t}$, de modo que $t=\dfrac{u^{2}}{2}$ y $dt=u\,du$. Entonces, cuando $t=0$ se tiene $u=0$, y cuando $t=x$ se tiene $u=\sqrt{2x}$. Por tanto,
$$
\int_{0}^{x} e^{\sqrt{2t}}\,dt=\int_{0}^{\sqrt{2x}} u e^{u}\,du
$$

Aplicando integración por partes,
$$
\int_{0}^{\sqrt{2x}} u e^{u}\,du=\left[u e^{u}\right]_{0}^{\sqrt{2x}}-\int_{0}^{\sqrt{2x}} e^{u}\,du
=\sqrt{2x}e^{\sqrt{2x}}-\left[e^{u}\right]_{0}^{\sqrt{2x}}
$$
y, por tanto,
$$
\int_{0}^{x} e^{\sqrt{2t}}\,dt=\sqrt{2x}e^{\sqrt{2x}}-e^{\sqrt{2x}}+1
$$

En consecuencia, una primitiva de $\int e^{\sqrt{2x}}\,dx$ es
$$
g(x)=\sqrt{2x}e^{\sqrt{2x}}-e^{\sqrt{2x}}
$$

### 2.

Una representación de tipo II del triángulo $T$ es
$$
T=\left\lbrace (x,y)\in\mathbb{R}^{2} : -1\leq y\leq 0,\ y\leq x\leq -y \right\rbrace
$$

Entonces, por integración reiterada,
$$
\int_{T} x\,dx\,dy=\int_{-1}^{0}\left(\int_{y}^{-y} x\,dx\right)\,dy
$$
pero, para cada $y\in[-1,0]$,
$$
\int_{y}^{-y} x\,dx=\left[\frac{x^{2}}{2}\right]_{y}^{-y}=\frac{(-y)^{2}-y^{2}}{2}=0
$$
y por tanto
$$
\int_{T} x\,dx\,dy=0
$$
