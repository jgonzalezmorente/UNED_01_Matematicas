>### Ejercicio 1
>
>Sean los siguientes conjuntos de $\mathbb{R}^3$
>$$
>\begin{aligned}
>A & = \left\lbrace \mathbf{a}_1, \mathbf{a}_2, \mathbf{a}_3 \right\rbrace
>   = \left\lbrace (0,0,1), (0,2,0), (3,0,0) \right\rbrace \\
>B & = \left\lbrace \mathbf{b}_1, \mathbf{b}_2, \mathbf{b}_3 \right\rbrace
>   = \left\lbrace (1,0,0), (1,1,0), (1,1,1) \right\rbrace
>\end{aligned}
>$$
>
>* Señale las matrices de cambio de base de $A$ a $B$ y de $B$ a $A$ respectivamente.
>* Sea $\mathbf{v}\in \mathbb{R}^3$ el vector de componentes $\mathbf{v}=(2,-2,1)$. Determine unas ecuaciones implícitas del subespacio generado por el conjunto dado $G[\{\mathbf{v}\}]$ en términos de las coordenadas respecto a $A$.

### Solución

Sea $E_3$ la base canónica de $\mathbb{R}^3$. Sabemos que las matrices de cambio de base de $A$ a $E_3$ y de $B$ a $E_3$ son
$$
M_{A\to E_3}=
\begin{pmatrix}
0 & 0 & 3 \\
0 & 2 & 0 \\
1 & 0 & 0
\end{pmatrix},
\quad
M_{B\to E_3}=
\begin{pmatrix}
1 & 1 & 1 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{pmatrix}
$$

Luego, las matrices de cambio de base de $A$ a $B$ y de $B$ a $A$ son
$$
M_{A\to B}=M_{B\to E_3}^{-1}M_{A\to E_3},
\qquad
M_{B\to A}=M_{A\to E_3}^{-1}M_{B\to E_3}
$$

Para calcular las inversas vamos a usar la fórmula
$$
M^{-1}=\frac{1}{\det(M)}\operatorname{Adj}(M)^T
$$
siendo $M$ una matriz cuadrada e inversible.

#### Inversa de $M_{A\to E_3}$

Calculamos en primer lugar el determinante
$$
\det(M_{A\to E_3})=-6
$$
y
$$
\operatorname{Adj}(M_{A\to E_3})^T=
\begin{pmatrix}
0 & 0 & -6 \\
0 & -3 & 0 \\
-2 & 0 & 0
\end{pmatrix}
$$
Luego,
$$
M_{A\to E_3}^{-1}
=
\frac{1}{-6}
\begin{pmatrix}
0 & 0 & -6 \\
0 & -3 & 0 \\
-2 & 0 & 0
\end{pmatrix}
=
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1/2 & 0 \\
1/3 & 0 & 0
\end{pmatrix}
$$

#### Inversa de $M_{B\to E_3}$

Calculamos el determinante
$$
\det(M_{B\to E_3})=1
$$
y
$$
\operatorname{Adj}(M_{B\to E_3})^T=
\begin{pmatrix}
1 & -1 & 0 \\
0 & 1 & -1 \\
0 & 0 & 1
\end{pmatrix}
$$
Por tanto,
$$
M_{B\to E_3}^{-1}
=
\frac{1}{1}
\begin{pmatrix}
1 & -1 & 0 \\
0 & 1 & -1 \\
0 & 0 & 1
\end{pmatrix}
=
\begin{pmatrix}
1 & -1 & 0 \\
0 & 1 & -1 \\
0 & 0 & 1
\end{pmatrix}
$$

Por tanto,
$$
M_{A\to B}
=
M_{B\to E_3}^{-1}M_{A\to E_3}
=
\begin{pmatrix}
1 & -1 & 0 \\
0 & 1 & -1 \\
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
0 & 0 & 3 \\
0 & 2 & 0 \\
1 & 0 & 0
\end{pmatrix}
=
\begin{pmatrix}
0 & -2 & 3 \\
-1 & 2 & 0 \\
1 & 0 & 0
\end{pmatrix}
$$

$$
M_{B\to A}
=
M_{A\to E_3}^{-1}M_{B\to E_3}
=
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1/2 & 0 \\
1/3 & 0 & 0
\end{pmatrix}
\begin{pmatrix}
1 & 1 & 1 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{pmatrix}
=
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1/2 & 1/2 \\
1/3 & 1/3 & 1/3
\end{pmatrix}
$$

Las coordenadas de $\mathbf{v}$ en la base $A$ se calculan aplicando la matriz de cambio de base de $E_3$ a $A$ al vector:
$$
\mathbf{v}_A=M_{A\to E_3}^{-1}\mathbf{v}
=
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1/2 & 0 \\
1/3 & 0 & 0
\end{pmatrix}
\begin{pmatrix}
2 \\
-2 \\
1
\end{pmatrix}
=
\begin{pmatrix}
1 \\
-1 \\
2/3
\end{pmatrix}
$$

Para determinar las ecuaciones implícitas, observemos que un vector de coordenadas $(x,y,z)$ pertenece a $G[\{\mathbf{v}\}]$ si y solo si
$$
\operatorname{rg}
\begin{pmatrix}
x & 1 \\
y & -1 \\
z & 2/3
\end{pmatrix}
=1
$$
y esto ocurre si y solo si todos los menores de orden $2$ tienen determinante nulo:
$$
\begin{aligned}
-x-y & = 0 \\
\frac{2}{3}y+z & = 0 \\
\frac{2}{3}x-z & = 0
\end{aligned}
$$

Dado que solo son necesarias dos ecuaciones independientes, una posible elección es
$$
\begin{aligned}
x+y & = 0 \\
2x-3z & = 0
\end{aligned}
$$

>### Ejercicio 2
>
>Sea $F:\mathbb{R}^2\to \mathbb{R}^2$ la aplicación lineal tal que
>$$
>\begin{aligned}
>F(1,0) &= (1,3) \\
>F(1,-1) &= (1,0)
>\end{aligned}
>$$
>
>* Calcule razonadamente el vector imagen $F(5,4)$.
>* Calcule la matriz $M(F;A,B)$ asociada a $F$ respecto de las bases
>$$
>A=\left\lbrace (1,1),(0,-2)\right\rbrace,\qquad 
>B=\left\lbrace (0,2),(-1,0)\right\rbrace
>$$

### Solución

Sea
$$
C=\left\lbrace 
\begin{pmatrix}1\\0\end{pmatrix},
\begin{pmatrix}1\\-1\end{pmatrix}
\right\rbrace
$$
que es base de $\mathbb{R}^2$ puesto que
$$
\det
\begin{pmatrix}
1 & 1\\
0 & -1
\end{pmatrix}
=-1\neq 0
$$

La matriz de $F$ respecto de las bases $C$ (en el dominio) y $E_2$ (base canónica en el codominio) es
$$
M(F;C,E_2)=
\begin{pmatrix}
1 & 1\\
3 & 0
\end{pmatrix}
$$
ya que sus columnas son las coordenadas en $E_2$ de $F(1,0)$ y $F(1,-1)$.

La matriz de cambio de base de $C$ a $E_2$ es
$$
M_{C\to E_2}=
\begin{pmatrix}
1 & 1\\
0 & -1
\end{pmatrix}
$$
y su inversa, que corresponde al cambio de base de $E_2$ a $C$, es
$$
M_{E_2\to C}=M_{C\to E_2}^{-1}=
\begin{pmatrix}
1 & 1\\
0 & -1
\end{pmatrix}
$$

Por composición, la matriz de $F$ respecto de la base canónica es
$$
M(F;E_2,E_2)=
M(F;C,E_2)\,M_{E_2\to C}=
\begin{pmatrix}
1 & 1\\
3 & 0
\end{pmatrix}
\begin{pmatrix}
1 & 1\\
0 & -1
\end{pmatrix}
=
\begin{pmatrix}
1 & 0\\
3 & 3
\end{pmatrix}
$$

En consecuencia,
$$
F(5,4)=
\begin{pmatrix}
1 & 0\\
3 & 3
\end{pmatrix}
\begin{pmatrix}
5\\
4
\end{pmatrix}
=
\begin{pmatrix}
5\\
27
\end{pmatrix}
$$

Para calcular $M(F;A,B)$ utilizamos
$$
M(F;A,B)=M_{E_2\to B}\,M(F;E_2,E_2)\,M_{A\to E_2}
$$

La matriz de cambio de base de $A$ a $E_2$ es
$$
M_{A\to E_2}=
\begin{pmatrix}
1 & 0\\
1 & -2
\end{pmatrix}
$$

La matriz de cambio de base de $B$ a $E_2$ es
$$
M_{B\to E_2}=
\begin{pmatrix}
0 & -1\\
2 & 0
\end{pmatrix}
$$
y, por tanto,
$$
M_{E_2\to B}=M_{B\to E_2}^{-1}=
\begin{pmatrix}
0 & 1/2\\
-1 & 0
\end{pmatrix}
$$

Finalmente,
$$
\begin{aligned}
M(F;A,B)
&=
\begin{pmatrix}
0 & 1/2\\
-1 & 0
\end{pmatrix}
\begin{pmatrix}
1 & 0\\
3 & 3
\end{pmatrix}
\begin{pmatrix}
1 & 0\\
1 & -2
\end{pmatrix}\\
&=
\begin{pmatrix}
3/2 & 3/2\\
-1 & 0
\end{pmatrix}
\begin{pmatrix}
1 & 0\\
1 & -2
\end{pmatrix}
=
\begin{pmatrix}
3 & -3\\
-1 & 0
\end{pmatrix}
\end{aligned}
$$

>### Ejercicio 3. 
>
>Este ejercicio consta de dos apartados, debe elegir y responder solamente a uno de ellos:
>
>1. En este apartado consideramos $f: \mathbb{R}\to \mathbb{R}$ la función
>$$
>f(x)=x \left| x - \frac{1}{2} \right|
>$$
>en donde $\left| \cdot \right|$ recordemos denota el valor absoluto. En este apartado se pide los siguiente:
>
>* Estudie la diferenciabilidad de $f$ en todo su dominio y calcule la derivada donde sea posible.
>* Calcule la integral
>$$
>\int_{-1}^{1} f(s) \, ds
>$$
>2. Sea $g:\mathbb{R}^2\to \mathbb{R}$ la función definida por
>$$
>g(x,y) = \int_{0}^{x-2y+2} (2s-1)^2 \, ds
>$$
>Calcule la derivada de $g$ en el punto $\mathbf{a}=(1,0)$ siguiendo el vector $\mathbf{v}=(1,-1)$.

### Solución

### 1.

Consideramos la función $f:\mathbb{R}\to\mathbb{R}$ definida por

$$
f(x)=x\left|x-\frac{1}{2}\right|
$$

Expresamos $f$ por ramas:
$$
f(x)=
\begin{cases}
x^2-\dfrac{x}{2}, & \text{si } x\geq \dfrac{1}{2} \\[0.3em]
-x^2+\dfrac{x}{2}, & \text{si } x\leq \dfrac{1}{2}
\end{cases}
$$

Como en cada rama $f$ es un polinomio, es derivable en todo punto $x\neq \dfrac{1}{2}$, y su derivada viene dada por
$$
f'(x)=
\begin{cases}
2x-\dfrac{1}{2}, & \text{si } x>\dfrac{1}{2} \\[0.3em]
-2x+\dfrac{1}{2}, & \text{si } x<\dfrac{1}{2}
\end{cases}
$$

En el punto $x=\dfrac{1}{2}$ calculamos las derivadas laterales:
$$
\begin{aligned}
f'\left(\frac{1}{2}\right)^- &= -2\cdot\frac{1}{2}+\frac{1}{2}=-\frac{1}{2} \\
f'\left(\frac{1}{2}\right)^+ &= 2\cdot\frac{1}{2}-\frac{1}{2}=\frac{1}{2}
\end{aligned}
$$
Como no coinciden, $f$ no es derivable en $x=\dfrac{1}{2}$.

Calculamos ahora la integral:
$$
\begin{aligned}
\int_{-1}^{1} f(s)\,ds
&=\int_{-1}^{1/2}\left(-s^2+\frac{s}{2}\right)\,ds
 +\int_{1/2}^{1}\left(s^2-\frac{s}{2}\right)\,ds \\
&=\left[-\frac{s^3}{3}+\frac{s^2}{4}\right]_{s=-1}^{s=1/2}
 +\left[\frac{s^3}{3}-\frac{s^2}{4}\right]_{s=1/2}^{s=1} \\
&=-\frac{11}{24}
\end{aligned}
$$

### 2.

Sea $g:\mathbb{R}^2\to\mathbb{R}$ definida por
$$
g(x,y)=\int_{0}^{x-2y+2}(2s-1)^2\,ds
$$

Definimos $h:\mathbb{R}^2\to\mathbb{R}$ por 

$$h(x,y)=x-2y+2$$

y $F:\mathbb{R}\to\mathbb{R}$ por
$$
F(x)=\int_{0}^{x}(2s-1)^2\,ds
$$
Entonces $g=F\circ h$.

Por el teorema fundamental del cálculo,
$$
F'(x)=(2x-1)^2
$$
y como $h(1,0)=3$, se tiene $F'(3)=25$. Además,
$$
h'(x,y)=\begin{pmatrix} 1 & -2 \end{pmatrix}
$$

Por la regla de la cadena,
$$
g'(1,0)=F'(h(1,0))\,h'(1,0)
=25\begin{pmatrix} 1 & -2 \end{pmatrix}
=\begin{pmatrix} 25 & -50 \end{pmatrix}
$$

La derivada direccional de $g$ en $\mathbf{a}=(1,0)$ según el vector $\mathbf{v}=(1,-1)$ es
$$
D_{\mathbf{v}}g(1,0)
=\begin{pmatrix} 25 & -50 \end{pmatrix}
\begin{pmatrix} 1 \\ -1 \end{pmatrix}
=75
$$

>### Ejercicio 4
>
>Este ejercicio consta de dos apartados; debe elegir y responder solamente a uno de ellos.
>
>1. En este apartado consideramos $f:\mathbb{R}^2\to\mathbb{R}$ la función definida por
>$$
>f(x,y)=\frac{1}{4}\left((1-x^2)^2+(1-y^2)^2\right)
>$$
>Se pide lo siguiente:
>- Calcule su función gradiente $\nabla f(x,y)$.
>- Determine todos sus puntos críticos.
>
>2. En este apartado consideramos la siguiente integral
>$$
>\int_{0}^{1}\int_{0}^{\sqrt{1-x^2}} \frac{x}{\sqrt{x^2+y^2}}\,dy\,dx
>$$
>Se pide lo siguiente:
>- Dibuje su recinto de integración y obtenga su expresión en coordenadas polares.
>- Calcule su valor.

### Solución

### 1

Calculemos las derivadas parciales de $f$:
$$
D_1 f(x,y)=\frac{1}{4}\,2(1-x^2)(-2x)=-x(1-x^2)
$$
$$
D_2 f(x,y)=\frac{1}{4}\,2(1-y^2)(-2y)=-y(1-y^2)
$$
Por tanto,
$$
\nabla f(x,y)=
\begin{pmatrix}
-x(1-x^2)\\
-y(1-y^2)
\end{pmatrix}
$$
Los puntos críticos son los puntos $(x,y)$ tales que $\nabla f(x,y)=0$, es decir,
$$
-x(1-x^2)=0
\quad\text{y}\quad
-y(1-y^2)=0
$$
Luego,
$$
x\in\left\lbrace -1,0,1\right\rbrace
\quad\text{y}\quad
y\in\left\lbrace -1,0,1\right\rbrace
$$
y los puntos críticos son
$$
\left\lbrace
(-1,-1),(-1,0),(-1,1),
(0,-1),(0,0),(0,1),
(1,-1),(1,0),(1,1)
\right\rbrace
$$

### 2

El recinto de integración viene dado por
$$
R=\left\lbrace (x,y):0\leq x\leq 1,\ 0\leq y\leq \sqrt{1-x^2}\right\rbrace
=
\left\lbrace (x,y):x\geq 0,\ y\geq 0,\ x^2+y^2\leq 1\right\rbrace
$$
En coordenadas polares,
$$
T=\left\lbrace (r,\theta):0\leq r\leq 1,\ 0\leq \theta\leq \frac{\pi}{2}\right\rbrace
=[0,1]\times\left[0,\frac{\pi}{2}\right]
$$
El cambio a coordenadas polares es $s:T\to R$ dado por
$$
s(r,\theta)=(r\cos\theta,r\sin\theta)
$$
El determinante del jacobiano es
$$
\det s'(r,\theta)=
\begin{vmatrix}
\cos\theta & -r\sin\theta\\
\sin\theta & r\cos\theta
\end{vmatrix}
=r
$$
Aplicando el teorema del cambio de variable,
$$
\int_{0}^{1}\int_{0}^{\sqrt{1-x^2}} \frac{x}{\sqrt{x^2+y^2}}\,dy\,dx
=
\int_{0}^{\pi/2}\int_{0}^{1}
\frac{r\cos\theta}{\sqrt{r^2}}\,r\,dr\,d\theta
=
\int_{0}^{\pi/2}\int_{0}^{1} r\cos\theta\,dr\,d\theta
$$
Finalmente,
$$
\int_{0}^{\pi/2}\int_{0}^{1} r\cos\theta\,dr\,d\theta
=
\int_{0}^{\pi/2}\cos\theta\left(\int_{0}^{1} r\,dr\right)d\theta
=
\frac{1}{2}
$$
