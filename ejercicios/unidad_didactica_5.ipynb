### Ejercicio 89

Resuelva las siguientes cuestiones:

1. Pruebe que $\mathbf{u}\cdot \mathbf{v} = 0$ si y sólo si $\|\mathbf{u} + \mathbf{v}\| = \| \mathbf{u} - \mathbf{v}\|$.
2. ¿Se encuentra el punto $(-1, 0, -1, 1)$ en la bola abierta de centro $(2, 1, -1, 2)$ y radio $\sqrt{3}$.
3. ¿Cuál es el coseno del ángulo que forman los vectores $(1,2,-1)$ y $(-1,1,0)$?

---

### Solución

1. Dado que
$$\|\mathbf{u} + \mathbf{v} \|^2= \|\mathbf{u}\|^2 + \|\mathbf{u}\|^2 + 2(\mathbf{u}\cdot \mathbf{v})$$
y 
$$\|\mathbf{u} - \mathbf{v} \|^2= \|\mathbf{u}\|^2 + \|\mathbf{u}\|^2 - 2(\mathbf{u}\cdot \mathbf{v})$$
se obtiene 
$$\|\mathbf{u} + \mathbf{v}\| = \| \mathbf{u} - \mathbf{v}\| \Leftrightarrow \|\mathbf{u} + \mathbf{v}\|^2 = \| \mathbf{u} - \mathbf{v}\|^2\Leftrightarrow \mathbf{u}\cdot \mathbf{v} = 0$$

2. Se trata de comprobar si la distancia del entre los dos puntos dados es menor o no al radio

```python
v1 = np.array([-1, 0, -1, 1])
v2 = np.array([2, 1, -1, 2])
distancia = np.linalg.norm(v1 - v2)
print(distancia < np.sqrt(3))
```

Por lo que el punto no pertenece a la bola.

3. Para calcular el coseno del ángulo que forman los vectores aplicamos
$$ \cos\theta = \frac{\mathbf{u}\cdot \mathbf{v}}{\lVert \mathbf{u} \rVert \lVert \mathbf{v} \rVert} $$

```python
u = np.array([1, 2, -1])
v = np.array([-1, 1, 0])
cos = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
print(cos)
```

### Ejercicio 90

Señale cuales de los siguientes conjuntos de $\mathbb{R}^2$ son acotados.

* $A=\left\lbrace \left( \dfrac{n+1}{n-1}, \dfrac{2n^2+1}{n^3-12} \right)\in \mathbb{R}^2: n=2,3,4,... \right\rbrace$
* $B=\left\lbrace (x,y,z)\in \mathbb{R}^3 : z = x^2 + y^2 \right\rbrace$
* $C=\left\lbrace (x,y)\in \mathbb{R}^2 : x + y < 3 \right\rbrace$
* $D=\left\lbrace (x,y)\in \mathbb{R}^2: x - y = 1 \right\rbrace$

---

### Solución

$A$ es acotado pues las sucesiones componentes que forman los puntos de $A$ son convergentes (la primera converge a $1$ y la segundoa a $0$) y por tanto acotadas. 

B$ no es acotado, si lo fuese existiría un $R>0$ tal que $C\subset B(0, R)$. Tomemos el punto $(\sqrt{R}, 0, R)\in C$, este punto no está en $B(0, R)$, ya que su norma es mayor a $R$,
$$
\lVert (\sqrt{R}, 0, R) \rVert = \sqrt{R + R^2} > R
$$

$C$ tampoco es acotado, si lo fuese estaría contenido en una bola $B(0, R)$ para algún $R>0$. Consideremos el punto $(0, -R - 1)$, este punto pertenece a $C$ pero no a $B(0, R)$, pues 
$$
\lVert (0, -R - 1) \rVert = R + 1 > R
$$

$D$ tampoco es acotado, los puntos de $D$ son de la forma $(x, 1-x)$ con $x\in \mathbb{R}$. La norma de un punto arbitrario de $D$ es por tanto
$$
\lVert (x, 1-x) \rVert = \sqrt{x^2 + (1-x)^2}
$$
que se puede hacer tan grande como se quiera tomando $x$ suficientemente grande.
