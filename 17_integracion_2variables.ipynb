# Introducción a la integración en dos variables

## Integración sobre un rectángulo $[a,b]\times[c,d]$

Este capítulo está dedicado a establecer las ideas básicas de cómo se extienden los conceptos de integración de una a varias variables. Por simplicidad, nos centraremos en funciones de dos variables. Al igual que en el capítulo anterior, a lo largo de este capítulo consideraremos siempre que
$$
f:\mathbb{R}^2\to\mathbb{R}
$$
es una función continua. Bajo esta hipótesis, la integral de $f$ sobre un rectángulo genérico del plano
$$
[a,b]\times[c,d]
=\left\lbrace (x,y)\in\mathbb{R}^2:\ a\leq x\leq b,\ c\leq y\leq d \right\rbrace
$$
que denotaremos por
$$
I=\int_{[a,b]\times[c,d]} f(x,y)\,dx\,dy
$$
existe.

Para definir un concepto de integral análogo al del caso de una variable, es necesario extender el concepto de partición de intervalos a rectángulos. Para ello, consideramos una sucesión finita de puntos equiespaciados en ambos ejes,
$$
x_i^n=a+\frac{i}{n}(b-a),\qquad
y_j^n=c+\frac{j}{n}(d-c),\qquad
i,j=0,\dots,n
$$
y los subintervalos en cada eje,
$$
I_i^n=[x_{i-1}^n,x_i^n]
=\left[a+\frac{i-1}{n}(b-a),\ a+\frac{i}{n}(b-a)\right]
$$
$$
J_j^n=[y_{j-1}^n,y_j^n]
=\left[c+\frac{j-1}{n}(d-c),\ c+\frac{j}{n}(d-c)\right]
$$
para $i=1,\dots,n$ y $j=1,\dots,n$.

Combinando todos los subintervalos se define una partición en $n^2$ rectángulos,
$$
R_{i,j}^n=I_i^n\times J_j^n
$$
todos ellos de igual área,
$$
\mathcal{A}\!\left(R_{i,j}^n\right)
=\mathcal{L}\!\left(I_i^n\right)\mathcal{L}\!\left(J_j^n\right)
=\frac{b-a}{n}\frac{d-c}{n}
=\frac{(b-a)(d-c)}{n^2}
$$

Sea el rectángulo $[a,b]\times[c,d]\subset\mathbb{R}^2$. Denotamos por $P_n^{[a,b]\times[c,d]}$ la partición equiespaciada de $[a,b]\times[c,d]$ dada por
$$
P_n^{[a,b]\times[c,d]}
=\left\lbrace R_{i,j}^n:\ i=1,\dots,n,\ j=1,\dots,n \right\rbrace
$$

---

### Fórmulas numéricas sobre rectángulos

Denotemos la integral doble sobre el rectángulo $[a,b]\times[c,d]$ por

$$
I=\int_{[a,b]\times[c,d]} f(x,y)dxdy
$$

Al igual que en el caso unidimensional, la propia definición de integral permite construir aproximaciones numéricas mediante sumas de Riemann. En efecto,

$$
\int_{[a,b]\times[c,d]} f(x,y)dxdy
=(b-a)(d-c)\lim_{n\to\infty}\frac{1}{n^2}\sum_{i=1}^{n}\sum_{j=1}^{n} f_{ij}
$$

de modo que, para $n$ suficientemente grande, la cantidad

$$
I_n=\frac{(b-a)(d-c)}{n^2}\sum_{i=1}^{n}\sum_{j=1}^{n} f_{ij}
$$

aproxima a $I$, donde

$$
f_{ij}=f!\left(a+\frac{i}{n}(b-a),,c+\frac{j}{n}(d-c)\right),\qquad i,j=1,\dots,n
$$

En particular, la sucesión ${I_n}$ converge a la integral, es decir, $I_n\to I$ cuando $n\to\infty$. Esto motiva la siguiente definición.

---

### Fórmula de un punto (compuesta) en dos variables

Definimos la **fórmula numérica de un punto de aplicación múltiple** en el rectángulo $[a,b]\times[c,d]$ como

$$
I_n\bigl(f;[a,b]\times[c,d]\bigr)=\frac{(b-a)(d-c)}{n^2}\sum_{i=1}^{n}\sum_{j=1}^{n} f_{ij}
$$

---

### Observación

Esta fórmula puede interpretarse como la aplicación reiterada de una regla “simple” sobre cada subrectángulo de una partición uniforme.

Sea

$$
x_i^n=a+\frac{i}{n}(b-a),\qquad y_j^n=c+\frac{j}{n}(d-c),\qquad i,j=0,\dots,n
$$

y consideremos los subrectángulos

$$
[x^n_{i-1},x^n_i]\times[y^n_{j-1},y^n_j],\qquad i,j=1,\dots,n
$$

Si aproximamos la integral en cada subrectángulo por el área del mismo multiplicada por el valor de $f$ en el vértice superior derecho, obtenemos

$$
\int_{[x^n_{i-1},x^n_i]\times[y^n_{j-1},y^n_j]} f(x,y)dxdy
\approx
\mathcal{A}\bigl([x^n_{i-1},x^n_i]\times[y^n_{j-1},y^n_j]\bigr)f(x_i^n,y_j^n)
$$

donde

$$
\mathcal{A}\bigl([x^n_{i-1},x^n_i]\times[y^n_{j-1},y^n_j]\bigr)=\frac{(b-a)(d-c)}{n^2}
$$

Sumando sobre todos los subrectángulos y usando la aditividad de la integral se llega a

\begin{equation*}
\begin{split}
\int_{[a,b]\times[c,d]} f(x,y)dxdy
&=\sum_{i=1}^{n}\sum_{j=1}^{n}\int_{[x^n_{i-1},x^n_i]\times[y^n_{j-1},y^n_j]} f(x,y)dxdy \\
&\approx \sum_{i=1}^{n}\sum_{j=1}^{n}\frac{(b-a)(d-c)}{n^2}f(x_i^n,y_j^n)
= \frac{(b-a)(d-c)}{n^2}\sum_{i=1}^{n}\sum_{j=1}^{n} f_{ij}
\end{split}
\end{equation*}

---

### Fórmula del trapecio compuesta en dos variables (por integración reiterada)

A partir de la fórmula compuesta del trapecio en una variable y usando **integración reiterada**, puede obtenerse una regla más precisa. Definimos

$$
g(x)=\int_c^d f(x,y)dy
$$

Entonces

$$
I=\int_a^b g(x)dx
$$

Aplicando la regla del trapecio compuesta a $g$ en $[a,b]$,

$$
I=\int_a^b g(x)dx \approx \frac{b-a}{2n}\left(g(x_0^n)+2\sum_{i=1}^{n-1}g(x_i^n)+g(x_n^n)\right)
$$

A su vez, cada término $g(x_i^n)$ es una integral en $y$, que aproximamos (de nuevo) por el trapecio compuesto en $[c,d]$:

$$
g(x_i^n)=\int_c^d f(x_i^n,y)dy
\approx
\frac{d-c}{2n}\left(f(x_i^n,y_0^n)+2\sum_{j=1}^{n-1}f(x_i^n,y_j^n)+f(x_i^n,y_n^n)\right)
$$

Con la notación

$$
f_{ij}=f(x_i^n,y_j^n),\qquad i,j=0,\dots,n
$$

esto se escribe como

$$
g(x_i^n)\approx \frac{d-c}{2n}\left(f_{i0}+2\sum_{j=1}^{n-1}f_{ij}+f_{in}\right),\qquad i=0,\dots,n
$$

Sustituyendo en la fórmula de $I$ y desarrollando, se obtiene finalmente

$$
I \approx I_n^1\bigl(f;[a,b]\times[c,d]\bigr)
$$

donde la **fórmula del trapecio compuesta en dos variables** viene dada por

$$
I_n^1\bigl(f;[a,b]\times[c,d]\bigr)
=\frac{(b-a)(d-c)}{4n^2}
\left[
\sum_{i=0}^{n}\sum_{j=0}^{n} w_i w_j f_{ij}
\right]
$$

con pesos

$$
w_0=w_n=1,\qquad w_i=2\ \text{si}\ i=1,\dots,n-1
$$

y análogamente para $w_j$.

Equivalente y más explícitamente, separando interior, bordes y vértices:

\begin{equation*}
\begin{split}
I_n^1\bigl(f;[a,b]\times[c,d]\bigr)
=\frac{(b-a)(d-c)}{4n^2}\Bigg[
& f_{00}+f_{0n}+f_{n0}+f_{nn} \\
& +2\sum_{j=1}^{n-1}(f_{0j}+f_{nj})
+2\sum_{i=1}^{n-1}(f_{i0}+f_{in}) \\
& +4\sum_{i=1}^{n-1}\sum_{j=1}^{n-1} f_{ij}
\Bigg]
\end{split}
\end{equation*}

---

### Notación del error

Denotaremos por $E_n$ el error asociado a la fórmula de un punto compuesta y por $E_n^1$ el error asociado al trapecio compuesto en dos variables:

$$
E_n\bigl(f;[a,b]\times[c,d]\bigr)=\int_{[a,b]\times[c,d]} f(x,y)dxdy - I_n\bigl(f;[a,b]\times[c,d]\bigr)
$$

$$
E_n^1\bigl(f;[a,b]\times[c,d]\bigr)=\int_{[a,b]\times[c,d]} f(x,y)dxdy - I_n^1\bigl(f;[a,b]\times[c,d]\bigr)
$$

## Integración sobre conjuntos medibles

### Conjuntos medibles

La teoría de la integración puede extenderse a conjuntos más generales que los rectángulos. Al igual que en el caso unidimensional, esto es posible para aquellos conjuntos a los que se les puede asociar un área. A dichos conjuntos se les denomina, en general, **conjuntos medibles** o **integrables**.

En $\mathbb{R}^2$, los conjuntos integrables que consideraremos pueden obtenerse mediante un número finito de uniones e intersecciones de recintos de integración que admiten dos tipos básicos de representación.

### Representaciones de conjuntos integrables

Diremos que un conjunto integrable $M\subset\mathbb{R}^2$ admite una **representación de tipo I** si puede escribirse como

$$
M={(x,y): a\leq x\leq b,\ g_1(x)\leq y\leq g_2(x)}
$$

donde $g_1,g_2:[a,b]\to\mathbb{R}$ son funciones continuas.

De manera análoga, diremos que un conjunto integrable admite una **representación de tipo II** si puede expresarse como

$$
M={(x,y): h_1(y)\leq x\leq h_2(y),\ c\leq y\leq d}
$$

donde en este caso $h_1,h_2:[c,d]\to\mathbb{R}$ son funciones continuas.

En general, un mismo conjunto $M$ puede admitir representaciones de ambos tipos. En tal situación, se escoge una u otra según convenga desde el punto de vista del cálculo de integrales. Veamos un ejemplo ilustrativo.

---

### Ejemplo 17.6

Consideremos el conjunto acotado $M$ limitado por la elipse

$$
\left(\frac{x}{2}\right)^2+y^2=1
$$

es decir,

$$
M=\left\lbrace (x,y): \left(\frac{x}{2}\right)^2+y^2\leq 1\right\rbrace
$$

Despejando en la desigualdad, obtenemos los límites para la variable $y$:

$$
\left(\frac{x}{2}\right)^2+y^2\leq 1
\Longleftrightarrow
y^2\leq 1-\left(\frac{x}{2}\right)^2
\Longleftrightarrow
-\sqrt{1-\left(\frac{x}{2}\right)^2}\leq y\leq \sqrt{1-\left(\frac{x}{2}\right)^2}
$$

Esta expresión solo tiene sentido cuando el radicando es no negativo, lo que determina los límites para la variable $x$:

$$
1-\left(\frac{x}{2}\right)^2\geq 0
\Longleftrightarrow
x^2\leq 4
\Longleftrightarrow
-2\leq x\leq 2
$$

Por tanto, el conjunto $M$ puede escribirse como

$$
M=\left\lbrace(x,y): -2\leq x\leq 2, -\sqrt{1-\left(\frac{x}{2}\right)^2}\leq y\leq \sqrt{1-\left(\frac{x}{2}\right)^2}\right\rbrace
$$

lo que proporciona una **representación de tipo I**, con

$$
a=-2,\qquad b=2,\qquad
g_1(x)=-\sqrt{1-\left(\frac{x}{2}\right)^2},\qquad
g_2(x)=\sqrt{1-\left(\frac{x}{2}\right)^2}
$$

Procediendo de forma análoga, podemos despejar ahora la variable $x$ en función de $y$, obteniendo

$$
M=\left\lbrace(x,y): -2\sqrt{1-y^2}\leq x\leq 2\sqrt{1-y^2},\ -1\leq y\leq 1\right\rbrace
$$

Esta expresión corresponde a una **representación de tipo II**, con

$$
c=-1,\qquad d=1,\qquad
h_1(y)=-2\sqrt{1-y^2},\qquad
h_2(y)=2\sqrt{1-y^2}
$$

---

### Cálculo de áreas mediante integración en dos variables

Por definición, los conjuntos $M$ que admiten representaciones de tipo I o de tipo II son conjuntos compactos, y su **área** puede calcularse mediante la integral doble

$$
\mathcal{A}(M)=\int_M dxdy
$$

donde $\mathcal{A}(M)$ denota el área del conjunto $M$. Obsérvese que

$$
\int_M dx,dy=\int_M 1dxdy
$$

representa el volumen del sólido cuya base es el conjunto $M$ y cuya altura es $1$. Dicho volumen es precisamente $\mathcal{A}(M)\cdot 1$, por lo que la integral doble $\int_M dxdy$ determina el área de $M$.

Utilizando las representaciones de tipo I y de tipo II, el cálculo del área puede realizarse mediante **integración reiterada**:

$$
\mathcal{A}(M)
=\int_a^b\int_{g_1(x)}^{g_2(x)} dydx
=\int_c^d\int_{h_1(y)}^{h_2(y)} dxdy
$$

---


### Ejemplo 17.7

Consideremos el conjunto

$$
M=\left\lbrace(x,y): -2\leq x\leq 2,\
-\sqrt{1-\left(\frac{x}{2}\right)^2}\leq y\leq \sqrt{1-\left(\frac{x}{2}\right)^2}\right\rbrace
$$

correspondiente a la elipse estudiada anteriormente. Su área viene dada por la integral

$$
\mathcal{A}(M)
=\int_{-2}^{2}\int_{-\sqrt{1-\left(\frac{x}{2}\right)^2}}^{\sqrt{1-\left(\frac{x}{2}\right)^2}} dydx
$$

Integrando primero respecto de $y$, obtenemos

$$
\mathcal{A}(M)
=2\int_{-2}^{2}\sqrt{1-\left(\frac{x}{2}\right)^2}dx
=\int_{-2}^{2}\sqrt{4-x^2}dx
$$

La integral anterior puede resolverse mediante el cambio de variable trigonométrico

$$
x(t)=2\sin t
$$

Calculamos la derivada y los nuevos límites de integración:

$$
x'(t)=2\cos t,\qquad
x\left(-\frac{\pi}{2}\right)=-2,\qquad
x\left(\frac{\pi}{2}\right)=2
$$

Aplicando la fórmula del cambio de variable, se obtiene

$$
\int_{-2}^{2}\sqrt{4-x^2}dx
=\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}
\sqrt{4-4\sin^2 t}2\cos tdt
=4\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\sqrt{1-\sin^2 t}\cos tdt
=4\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 tdt
$$

Por tanto,

$$
\mathcal{A}(M)=4\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 t dt
$$


Para calcular la integral restante, aplicamos integración por partes. Recordemos la fórmula:

$$
\int_a^b v'(t)u(t)dt
= v(b)u(b)-v(a)u(a)-\int_a^b v(t)u'(t)dt
$$

Tomando

$$
u(t)=\cos t,\qquad v'(t)=\cos t,
$$

se tiene $u'(t)=-\sin t$ y $v(t)=\sin t$. Aplicando la fórmula, resulta

\begin{equation*}
\begin{split}
\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 tdt
&= \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\sin t\sin tdt
= \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\sin^2 tdt \\
&= \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(1-\cos^2 t),dt
= \pi-\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 t dt
\end{split}
\end{equation*}

Despejando, se obtiene

$$
\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 t dt=\frac{\pi}{2}
$$

Finalmente,

$$
\mathcal{A}(M)
=4\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}\cos^2 t dt
=2\pi
$$

### Ejemplo 17.8

Calcular el área del triángulo $T$ de vértices

$$
\{(1,1),\ (2,0),\ (2,2)\}
$$

Por definición, el área de un conjunto plano $T$ viene dada por la integral doble

$$
\mathcal{A}(T)=\int_T dxdy
$$

Geométricamente, el triángulo $T$ admite una **representación de tipo I**, que puede describirse como

$$
T=\{(x,y): 1\leq x\leq 2,\ 2-x\leq y\leq x\}
$$

En efecto, para cada valor de $x\in[1,2]$, la variable $y$ queda acotada entre las rectas

$$
y=2-x \qquad \text{y} \qquad y=x
$$

que se cortan en el punto $(1,1)$ y alcanzan el segmento vertical $x=2$ en los puntos $(2,0)$ y $(2,2)$, respectivamente.

Aplicando integración reiterada, se obtiene

$$
\mathcal{A}(T)
=\int_{1}^{2}\left(\int_{2-x}^{x} dy\right)dx
=\int_{1}^{2}(x-(2-x))dx
=\int_{1}^{2}(2x-2)dx
$$

Calculando la integral exterior,

$$
\mathcal{A}(T)
=\left[x^2-2x\right]_{1}^{2}
=(4-4)-(1-2)
=1
$$

Este resultado coincide con el cálculo geométrico directo: el triángulo $T$ tiene base $2$ y altura $1$, por lo que su área es

$$
\mathcal{A}(T)=\frac{1}{2}\cdot 2\cdot 1=1
$$

Esto confirma la coherencia entre la interpretación geométrica y el cálculo mediante integración doble.

## Integrales sobre conjuntos medibles

Sea $f:\mathbb{R}^2\to\mathbb{R}$ una función continua y sea $M\subset\mathbb{R}^2$ un conjunto medible. Podemos definir la **integral de $f$ sobre $M$**, que se denota por

$$
\int_M f(x,y)dxdy
$$

### Cálculo mediante representaciones del dominio

Si el conjunto $M$ admite una **representación de tipo I**,

$$
M=\{(x,y): a\leq x\leq b,\ g_1(x)\leq y\leq g_2(x)\}
$$

la integral de $f$ sobre $M$ se calcula mediante integración reiterada como

$$
\int_M f(x,y)dxdy
=\int_a^b\left(\int_{g_1(x)}^{g_2(x)} f(x,y)dy\right)dx
$$

De forma análoga, si $M$ admite una **representación de tipo II**,

$$
M=\{(x,y): c\leq y\leq d,\ h_1(y)\leq x\leq h_2(y)\}
$$

la integral de $f$ sobre $M$ viene dada por

$$
\int_M f(x,y)dxdy
=\int_c^d\left(\int_{h_1(y)}^{h_2(y)} f(x,y)dx\right)dy
$$

Como se ha visto anteriormente, para un mismo conjunto $M$ pueden existir representaciones de tipo I y de tipo II. La elección de una u otra puede simplificar notablemente el cálculo de la integral, dependiendo de la forma del dominio y de la expresión de la función integranda.

### Propiedades básicas de la integral doble

La integral en dos variables comparte propiedades fundamentales con la integral de una variable, como la **linealidad**, la **aditividad respecto del dominio** y la **positividad**. Estas propiedades permiten, además, descomponer dominios complicados en otros más sencillos y simplificar el cálculo de integrales.

### Proposición 17.2

Sean $M,N\subset\mathbb{R}^2$ conjuntos medibles y sean $f,g:\mathbb{R}^2\to\mathbb{R}$ funciones continuas. Entonces se verifican las siguientes propiedades:

* **Linealidad.** Para cualesquiera $\alpha,\beta\in\mathbb{R}$ se tiene

  $$
  \int_M \bigl(\alpha f(x,y)+\beta g(x,y)\bigr)dxdy
  =\alpha\int_M f(x,y),dx,dy+\beta\int_M g(x,y)dxdy
  $$

* **Aditividad respecto del dominio.**
  Si $M\cap N$ es un conjunto de área nula (por ejemplo, si es vacío, una recta o una curva), entonces

  $$
  \int_{M\cup N} f(x,y),dx,dy
  =\int_M f(x,y),dx,dy+\int_N f(x,y)dxdy
  $$

* **Positividad.**
  Si

  $$
  f(x,y)\geq 0 \qquad \text{para todo } (x,y)\in M
  $$

  entonces

  $$
  \int_M f(x,y)dxdy\geq 0
  $$

### Ejemplo 17.9

Calcular la integral

$$
\int_M x^3 ydxdy
$$

donde $M$ es el recinto limitado por la parte positiva de los ejes coordenados, la circunferencia

$$
x^2+y^2=1
$$

y la elipse

$$
x^2+4y^2=4
$$

Geométricamente, el recinto de integración $M$ admite una **representación de tipo II**, que puede escribirse como

$$
M=\left\lbrace(x,y): \sqrt{1-y^2}\leq x\leq 2\sqrt{1-y^2},\ 0\leq y\leq 1\right\rbrace
$$

Aplicando integración reiterada, se obtiene

\begin{equation*}
\begin{split}
\int_M x^3 ydxdy
&= \int_0^1\left(\int_{\sqrt{1-y^2}}^{2\sqrt{1-y^2}} x^3 ydx\right)dy \\
&= \int_0^1 y\left(\int_{\sqrt{1-y^2}}^{2\sqrt{1-y^2}} x^3dx\right)dy
\end{split}
\end{equation*}

Calculando la integral interior,

$$
\int_{\sqrt{1-y^2}}^{2\sqrt{1-y^2}} x^3dx
=\frac{1}{4}\left[16(1-y^2)^2-(1-y^2)^2\right]
=\frac{15}{4}(1-y^2)^2
$$

Por tanto,

\begin{equation*}
\begin{split}
\int_M x^3 ydxdy
&= \frac{15}{4}\int_0^1 y(1-y^2)^2dy \\
&= \frac{15}{4}\int_0^1 (y^5-2y^3+y)dy \\
&= \frac{15}{4}\left(\frac{1}{6}-\frac{1}{2}+\frac{1}{2}\right)
=\frac{5}{8}
\end{split}
\end{equation*}

Alternativamente, podemos descomponer el recinto $M$ como unión de dos conjuntos de **tipo I**,

$$
M=M_1\cup M_2,
$$

donde

$$
M_1=\left\lbrace(x,y): 0\leq x\leq 1,\ \sqrt{1-x^2}\leq y\leq \sqrt{1-\frac{x^2}{4}}\right\rbrace
$$

$$
M_2=\left\lbrace(x,y): 1\leq x\leq 2,\ 0\leq y\leq \sqrt{1-\frac{x^2}{4}}\right\rbrace
$$

La intersección $M_1\cap M_2$ es un segmento, y por tanto un conjunto de área nula. En consecuencia, podemos aplicar la propiedad de **aditividad** de la integral y escribir

$$
\int_M x^3 ydxdy
=\int_{M_1} x^3 ydxdy+\int_{M_2} x^3 ydxdy
$$

\begin{equation*}
\begin{split}
\int_{M_1} x^3 ydxdy
&= \int_0^1\left(\int_{\sqrt{1-x^2}}^{\sqrt{1-\frac{x^2}{4}}} x^3 ydy\right)dx \\
&= \int_0^1 x^3\left(\int_{\sqrt{1-x^2}}^{\sqrt{1-\frac{x^2}{4}}} ydy\right)dx \\
&= \int_0^1 \frac{x^3}{2}\left(1-\frac{x^2}{4}-1+x^2\right)dx \\
&= \frac{3}{8}\int_0^1 x^5dx
=\frac{3}{8}\cdot\frac{1}{6}
=\frac{1}{16}
\end{split}
\end{equation*}

\begin{equation*}
\begin{split}
\int_{M_2} x^3 ydxdy
&= \int_1^2\left(\int_0^{\sqrt{1-\frac{x^2}{4}}} x^3 ydy\right)dx \\
&= \int_1^2 x^3\left(\int_0^{\sqrt{1-\frac{x^2}{4}}} ydy\right)dx \\
&= \int_1^2 \frac{x^3}{2}\left(1-\frac{x^2}{4}\right)dx \\
&= \int_1^2 \frac{x^3}{2}dx - \int_1^2 \frac{x^5}{8}dx \\
&= \left(2-\frac{1}{8}\right)-\left(\frac{4}{3}-\frac{1}{48}\right)
=\frac{9}{16}
\end{split}
\end{equation*}

Sumando ambas contribuciones,

$$
\int_M x^3 ydxdy
=\frac{1}{16}+\frac{9}{16}
=\frac{5}{8}
$$

que coincide con el resultado obtenido anteriormente mediante la representación de tipo II.

Este ejemplo ilustra claramente cómo **la elección adecuada del dominio de integración** puede simplificar de forma notable el cálculo de una integral doble.

## Cambios de variable en el plano

En esta sección se generaliza la noción de **cambio de variable** del caso de una variable al caso de dos variables. Recordemos, en primer lugar, el teorema del cambio de variable en una dimensión.

Sea $f:[a,b]\to\mathbb{R}$ una función continua y sea $x:[c,d]\to\mathbb{R}$ una función derivable con derivada continua tal que

$$
x(c)=a,\qquad x(d)=b
$$

Entonces se verifica

$$
\int_{x([c,d])=[a,b]} f(s)ds=
\int_{c}^{d} f(x(t))|x'(t)|dt
$$

De manera informal, se dice que la función $x:[c,d]\to[a,b]$ define un **cambio de variable en $\mathbb{R}$**, ya que permite transformar la integral sobre el intervalo $[a,b]$ en una integral sobre $[c,d]$.

En el caso del plano $\mathbb{R}^2$, la idea es análoga: se pretende transformar una integral sobre un conjunto $M$ en otra integral sobre un conjunto $T$ mediante una aplicación suficientemente regular. A continuación se formaliza esta noción.

### Definición 17.2

Una función $s:\mathbb{R}^2\to\mathbb{R}^2$, dada por

$$
s(t_1,t_2)=\bigl(s_1(t_1,t_2),,s_2(t_1,t_2)\bigr),
$$

define un **cambio de variable** entre dos conjuntos $T$ y $M$ de $\mathbb{R}^2$ si se cumplen las siguientes condiciones:

* **Correspondencia de dominios.**
  $$
  s(T)={s(t): t\in T}=M
  $$

* **No degeneración local.**
  El determinante de la matriz jacobiana de $s$ no se anula en ningún punto de $T$, es decir,

  $$
  \det s'(t)
  =
  \begin{vmatrix}
  D_1 s_1(t) & D_2 s_1(t) \\
  D_1 s_2(t) & D_2 s_2(t)
  \end{vmatrix}
  \neq 0,
  \qquad \text{para todo } t=(t_1,t_2)\in T.
  $$

Esta última condición garantiza que la transformación $s$ es localmente invertible y que no colapsa áreas, lo cual será esencial para poder expresar integrales sobre $M$ en términos de integrales sobre $T$ mediante el **determinante jacobiano**.

## Cambio a coordenadas polares

Uno de los cambios de variable más importantes es el cambio a coordenadas polares. Dado un punto de coordenadas cartesianas $(x,y)$, sus coordenadas polares son un par $(r,\theta)$, con $r\geq 0$ y $\theta\in[0,2\pi)$, que verifican
$$
x=r\cos\theta,\qquad y=r\sin\theta
$$
La interpretación geométrica es inmediata: $r$ representa el módulo del vector de posición y $\theta\in[0,2\pi)$ el ángulo que dicho vector forma con el eje de abscisas (eje $X$), medido en sentido contrario a las agujas del reloj.

Por ejemplo, al punto de coordenadas cartesianas
$$
(x,y)=(1,1)
$$
le corresponden las coordenadas polares
$$
\left(\sqrt{2},\frac{\pi}{4}\right)
$$

Formalmente, dado un punto de coordenadas cartesianas $(x,y)\in\mathbb{R}^2$, sus coordenadas polares se obtienen mediante las fórmulas
$$
r=\sqrt{x^2+y^2}
$$
y
$$
\theta=\begin{cases}
\arctan\left(\dfrac{y}{x}\right)       & \text{si } x>0,\ y\geq 0 \\
\dfrac{\pi}{2}                         & \text{si } x=0,\ y>0     \\
\arctan\left(\dfrac{y}{x}\right)+\pi   & \text{si } x<0           \\
\dfrac{3\pi}{2}                        & \text{si } x=0,\ y<0     \\
\arctan\left(\dfrac{y}{x}\right)+2\pi  & \text{si } x>0,\ y<0
\end{cases}
$$
donde se tiene en cuenta que la función $\arctan$ toma valores en el intervalo $\left(-\dfrac{\pi}{2},\dfrac{\pi}{2}\right)$ y se adopta el convenio $\theta\in[0,2\pi)$. En la práctica, la determinación del ángulo suele realizarse de manera más sencilla mediante una consideración geométrica que aplicando directamente la expresión analítica.

Para el punto $(x,y)=(0,0)$ se toma $r=0$, quedando el ángulo indefinido, lo cual no supone ninguna dificultad al no ser necesario asignarle un valor.

Desde un punto de vista formal, las ecuaciones anteriores definen un cambio de variable entre los conjuntos
$$
T=\{(r,\theta): r>0,\ 0\leq \theta<2\pi\}
$$
y
$$
M=s(T)=\{(x,y): (x,y)\neq(0,0)\},
$$
donde la aplicación de cambio de variable viene dada por
$$
s:(0,\infty)\times[0,2\pi)\longrightarrow \mathbb{R}^2\setminus{(0,0)},\qquad
s(r,\theta)=(r\cos\theta,r\sin\theta)
$$
El determinante de la matriz jacobiana de $s$ viene dado por
$$
\det s'(r,\theta)
=\begin{vmatrix}
\cos\theta & -r\sin\theta \\
\sin\theta & r\cos\theta
\end{vmatrix}
=r
$$
que no se anula para ningún $(r,\theta)\in T$. Por tanto, estrictamente hablando, $s$ no define un cambio de variable en conjuntos que incluyan puntos donde $r=0$, es decir, el origen de coordenadas $(0,0)$. No obstante, en el cálculo de integrales esto no tiene relevancia práctica, ya que el conjunto $\{(0,0)\}$ tiene área nula y puede incluirse sin afectar al resultado.

El uso de coordenadas polares simplifica notablemente la representación de ciertos conjuntos, como circunferencias, elipses o coronas circulares, y, como se verá más adelante, facilita el cálculo de integrales sobre dichos dominios. A continuación se presentan algunos ejemplos ilustrativos.

---

### Ejemplo 17.10

Consideremos el círculo de radio $2$ y centro en el origen $(0,0)$. En coordenadas cartesianas, dicho conjunto admite una representación de tipo II dada por
$$
M=\left\lbrace (x,y): -2\leq y \leq 2,\ -\sqrt{4-y^2}\leq x \leq \sqrt{4-y^2} \right\rbrace
$$

Al pasar a coordenadas polares, el conjunto $M$ se transforma en el rectángulo
$$
R=[0,2]\times [0,2\pi)
$$

Sea ahora $N$ la parte del primer cuadrante del círculo $M$. Este conjunto puede describirse en coordenadas cartesianas mediante la representación de tipo II
$$
N=\left\lbrace (x,y): 0\leq y \leq 2,\ 0\leq x \leq \sqrt{4-y^2} \right\rbrace
$$

En coordenadas polares, el conjunto $N$ se corresponde con el subrectángulo
$$
L=[0,2]\times \left[0,\frac{\pi}{2}\right]
$$

### Ejemplo 17.11

Consideremos el conjunto $M$ dado en coordenadas cartesianas como la porción de corona circular delimitada por las circunferencias de radios $1$ y $2$ y por las rectas $y=x$ e $y=-x$. Este conjunto admite una representación de tipo I analíticamente complicada:
$$
M=\left\lbrace (x,y): -\sqrt{2}\leq x \leq \sqrt{2}, \ g_1(x) \leq y \leq g_2(x)\right\rbrace

donde

g_1(x)=\begin{cases}

\sqrt{1-x^2} & \text{si } -\dfrac{\sqrt{2}}{2}\leq x\leq\dfrac{\sqrt{2}}{2} \\
x            & \text{si } \dfrac{\sqrt{2}}{2}\leq x\leq\sqrt{2}
\end{cases}
$$
y
$$
g_2(x)=\sqrt{4-x^2}
$$
En cambio, como se observa, dicha subcorona se corresponde con un rectángulo en coordenadas polares:
$$
R=[1,2]\times\left[\frac{\pi}{4}, \frac{3\pi}{4}\right]
$$

## Transformaciones afines entre triángulos

Las transformaciones afines entre triángulos son aplicaciones afines que transforman un triángulo en otro. Como veremos, constituyen cambios de variable en el plano. Tienen una gran importancia en matemáticas por sus numerosas aplicaciones y, en nuestro contexto, proporcionan un ejemplo fundamental de aplicación afín y de cambio de variable.

Recordemos que una aplicación afín de $\mathbb{R}^n$ en $\mathbb{R}^m$ es una aplicación lineal a la que se le suma un vector constante, es decir, una aplicación $F:\mathbb{R}^n\to\mathbb{R}^m$ del tipo
$$
F(x)=Bx+c
$$
donde $B\in\mathcal{M}_{m\times n}$ y $c\in\mathbb{R}^m$ son, respectivamente, una matriz y un vector constante.

Sea ahora $T$ un triángulo de $\mathbb{R}^2$ definido por los vértices
$$
\{a_1=(a_{11},a_{21}),\ a_2=(a_{12},a_{22}),\ a_3=(a_{13},a_{23})\}
$$
suponiendo que dichos puntos no están alineados. Sea $\hat{T}$ el triángulo de referencia definido por los vértices
$$
\{\hat{a}_1=(0,0),\ \hat{a}_2=(1,0),\ \hat{a}_3=(0,1)\}
$$

Denotamos por $s_\triangle:\mathbb{R}^2\to\mathbb{R}^2$ la aplicación definida por
$$
s_\triangle
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=
\begin{pmatrix}
a_{12}-a_{11} & a_{13}-a_{11} \\
a_{22}-a_{21} & a_{23}-a_{21}
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
+
\begin{pmatrix}
a_{11} \\
a_{21}
\end{pmatrix}
$$
para todo $x=(x_1,x_2)\in\mathbb{R}^2$.

Tenemos el siguiente resultado.

### Proposición 17.3

Se verifican las siguientes propiedades:

* $s_\triangle$ es una aplicación afín que hace corresponder a cada vértice de $\hat{T}$ un vértice de $T$. En particular,
  $$
  s_\triangle(\hat{a}_i)=a_i \quad \text{para } i=1,2,3
  $$

* La aplicación $s_\triangle:\hat{T}\to T$ es un cambio de variable entre los triángulos $\hat{T}$ y $T$. En efecto:

  * $s_\triangle(\hat{T})=T$
  * $\det s'_\triangle(x)=\det\begin{pmatrix}
        a_{12}-a_{11} & a_{13}-a_{11} \\
        a_{22}-a_{21} & a_{23}-a_{21}
    \end{pmatrix}\neq 0$

---

### Ejemplo 17.12

Sea $T$ el triángulo de vértices
$$
\{a_1=(1,1),\ a_2=(2,0),\ a_3=(2,2)\}
$$
En este caso, la aplicación afín $s_\triangle$ viene dada por
$$
s_\triangle
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
+
\begin{pmatrix}
1 \\
1
\end{pmatrix}
$$
Los puntos no están alineados, ya que el determinante de la matriz asociada a la parte lineal de la aplicación afín es no nulo,
$$
\begin{vmatrix}
1 & 1 \\
-1 & 1
\end{vmatrix}
=
2
$$
La aplicación $s_\triangle$ hace corresponder cada vértice del triángulo de referencia $\hat{T}$ con un vértice de $T$. Del mismo modo, cada punto de $\hat{T}$ se corresponde con un único punto de $T$, ya que la aplicación $s_\triangle$ determina una biyección entre ambos triángulos.

## Teorema de cambio de variable en el plano

Una vez que se dispone de una noción precisa de cambio de variable, puede establecerse la generalización del teorema de cambio de variable para el caso de dos variables.

### Teorema 17.1

**Teorema de cambio de variable en $\mathbb{R}^2$**.
Sea $f:\mathbb{R}^2\to\mathbb{R}$ una función continua y sea $s:\mathbb{R}^2\to\mathbb{R}^2$ un cambio de variable entre dos conjuntos integrables $T$ y $M$ de $\mathbb{R}^2$. Entonces se cumple que
$$
\int_{M=s(T)} f(x_1,x_2) dx_1dx_2
=
\int_T f\bigl(s_1(t_1,t_2), s_2(t_1,t_2)\bigr)|\det s'(t_1,t_2)|dt_1dt_2
$$

#### Observación 17.3

Si se considera la función constante $f=1$, se obtiene una fórmula para calcular el área del conjunto $M$,
$$
\mathcal{A}(M)
=\int_M dx_1,dx_2
=\int_T |\det s'(t_1,t_2)|dt_1dt_2
$$
Por tanto, la función escalar definida mediante el determinante del jacobiano del cambio de variable,
$$
g(t_1,t_2)=|\det s'(t_1,t_2)|
$$
es un factor positivo que determina, en cierto sentido, cómo varían las áreas de los conjuntos $T$ y $M$. De manera informal, puede escribirse
$$
dx_1dx_2=|\det s'(t_1,t_2)|dt_1dt_2
$$
lo que relaciona las áreas de los elementos de las particiones de $M$ y $T$, respectivamente.

Por ejemplo, en el cambio a coordenadas polares, como el determinante del jacobiano viene dado por el radio $r$, se obtiene
$$
dxdy = rdrd\theta
$$
En el caso de transformaciones afines entre triángulos, se obtiene la fórmula
$$
\mathcal{A}(T)=\int_T dxdy = \int_{\hat{T}} |\det s'(t)|dt_1dt_2
$$
Como
$$
\det s'(t)=\det (a_2-a_1, a_3-a_1)
$$
es constante con respecto a la variable $t$, puede extraerse fuera de la integral, obteniéndose
$$
\mathcal{A}(T)=|\det (a_2-a_1, a_3-a_1)|=\int_{\hat{T}} dt_1dt_2=|\det (a_2-a_1, a_3-a_1)|\mathcal{A}(\hat{T})
$$
El área del triángulo de referencia es directamente calculable,
$$
\mathcal{A}(\hat{T})=\frac{1}{2}
$$
y, por tanto,
$$
\mathcal{A}(T)=\frac{1}{2}|\det (a_2-a_1, a_3-a_1)|
$$
lo que coincide con la fórmula conocida. Obsérvese que dicha área es nula si el determinante es cero y, en consecuencia, los puntos $a_1$, $a_2$ y $a_3$ están alineados.

### Ejemplo 17.13

Sea de nuevo el triángulo $T$ de vértices
$$
\{(1,1),\ (2,0),\ (2,2)\}
$$
Como se vio en el ejemplo anterior, el cambio de variable afín que envía el triángulo de referencia sobre $T$ viene dado por
$$
s_\triangle(x)=
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
+
\begin{pmatrix}
1 \\
1
\end{pmatrix}
$$
El área de $T$ se obtiene como la mitad del valor absoluto del determinante del jacobiano de dicho cambio, es decir,
$$
\mathcal{A}(T)=\frac{1}{2}\left|\det s_\triangle'(x)\right|
=\frac{1}{2}
\begin{vmatrix}
1 & 1 \\
-1 & 1
\end{vmatrix}
=1
$$
lo cual coincide con el resultado obtenido en el Ejemplo 17.8.

### Ejemplo 17.14

Utilice el cambio de variable a polares para calcular el área del recinto elíptico
$$
M=\left\lbrace (x,y): \left(\frac{x}{2}\right)^2+y^2\leq 1\right\rbrace
$$

Consideremos, en general, el recinto elíptico
$$
E=\left\lbrace (x,y): \left(\frac{x}{a}\right)^2+\left(\frac{y}{b}\right)^2\leq 1\right\rbrace
$$
donde
$$
\left\lbrace (x,y): \left(\frac{x}{a}\right)^2+\left(\frac{y}{b}\right)^2=1\right\rbrace
$$
es la expresión analítica de una elipse de parámetros $a,b>0$.

Mediante el cambio de variable
$$
X=\frac{x}{a},\qquad Y=\frac{y}{b}
$$
se transforma la elipse en el disco unidad
$$
X^2+Y^2\leq 1
$$
cuya frontera viene dada por $X^2+Y^2=1$. En coordenadas polares,
$$
X=r\cos\theta,\qquad Y=r\sin\theta
$$
el dominio correspondiente es
$$
R=\left\lbrace (r,\theta): 0\leq r\leq 1,\ 0\leq \theta<2\pi\right\rbrace
$$

Deshaciendo los cambios, se obtiene la aplicación $s:\mathbb{R}^2\to\mathbb{R}^2$ dada por
$$
\begin{aligned}
x&=s_1(r,\theta)=ar\cos\theta\\
y&=s_2(r,\theta)=br\sin\theta
\end{aligned}
$$
de modo que $s(R)=E$.

El determinante del jacobiano es
$$
\det s'(r,\theta)=
\begin{vmatrix}
a\cos\theta & -ar\sin\theta\\
b\sin\theta & br\cos\theta
\end{vmatrix}
=abr
$$
y, en particular, $\det s'(r,\theta)>0$ si $r>0$.

Aplicando la fórmula del cambio de variable,
$$
\mathcal{A}(E)=\int_R \left|\det s'(r,\theta)\right|\,dr\,d\theta
=\int_0^1\left(\int_0^{2\pi}abr\,d\theta\right)dr
=\pi ab
$$

Para el caso particular $a=2$ y $b=1$, se obtiene
$$
\mathcal{A}(M)=2\pi
$$

### Ejemplo 17.15

Calcule la integral
$$
I=\int_M e^{-(x^2+y^2)}\,dx\,dy
$$
donde el recinto de integración $M$ es el círculo de centro $(0,0)$ y radio $2$.

Una representación de tipo II de $M$ es
$$
M=\left\lbrace (x,y)\in\mathbb{R}^2:\ -2\leq y\leq 2,\ -\sqrt{4-y^2}\leq x\leq \sqrt{4-y^2}\right\rbrace
$$

Consideramos el cambio a coordenadas polares
$$
\begin{aligned}
x&=r\cos\theta\\
y&=r\sin\theta
\end{aligned}
$$
cuyo dominio en el plano $(r,\theta)$ es
$$
T=[0,2]\times[0,2\pi)
$$
y cuyo jacobiano verifica
$$
dx\,dy=r\,dr\,d\theta
$$

Por tanto,
$$
\begin{aligned}
I
&=\int_{-2}^2\left(\int_{-\sqrt{4-y^2}}^{\sqrt{4-y^2}} e^{-(x^2+y^2)}\,dx\right)\,dy\\
&=\int_0^2\left(\int_0^{2\pi} e^{-r^2}\,r\,d\theta\right)\,dr\\
&=\int_0^2 2\pi r e^{-r^2}\,dr\\
&=\pi\left(1-e^{-4}\right)
\end{aligned}
$$
