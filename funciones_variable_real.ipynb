## El conjunto de los números reales. Sucesiones

>**Proposición 8.3 (Unicidad del límite)**
>
>El límite de una sucesión, si existe, es único. Además, cualquier subsucesión de una sucesión convergente tiende al mismo límite.

>**Proposición 8.4**
>
>Sean $\{a_n\}, \ \{b_n\}\subset\mathbb{R}$ y $\lambda\in\mathbb{R}$. Si $\{a_n\}\to a\in\overline{\mathbb{R}}$ y $\{b_n\}\to b\in\overline{\mathbb{R}}$, entonces:
>
>1. $\lim\limits_{n\to\infty} (a_n + b_n) = a + b$
>
>2. $\lim\limits_{n\to\infty} \lambda a_n = \lambda a$
>
>3. $\lim\limits_{n\to\infty} a_nb_n = ab$
>
>4. Si $b_n\neq 0$ para todo $n\in\mathbb{N}$, entonces $\lim\limits_{n\to\infty}\dfrac{a_n}{b_n}=\dfrac{a}{b}$
>
>5. Si $\{a_n\}$ es de términos positivos, entonces $\lim\limits_{n\to\infty} a_n^{b_n} = a^b$
>
>siempre y cuando en lo anterior no se dé lugar a ninguna indeterminación.

>**Proposición 8.5**
>
>Toda sucesión convergente es acotada.

>**Proposición 8.6**
>
>Toda sucesión $\{a_n\}$ monótona y acotada es convergente.

>**Proposición 8.7**
>
>Si $\lim\limits_{n\to\infty} a_n=\pm\infty$, entonces 
>$$\lim_{n\to\infty}\left(1+\dfrac{1}{a_n}\right)^{a_n}=e$$

>**Proposición 8.8**
>
>Sean $\{a_n\}, \ \{b_n\}\subset\mathbb{R}$, donde $\{a_n\}$ es una sucesión de términos positivos. Se tiene que
>$$\lim_{n\to\infty}a_n^{b_n}=\lim_{n\to\infty}e^{\ln \left(a_n^{b_n}\right)}=\lim_{n\to\infty}e^{b_n\ln(a_n)}$$

>**Proposición 8.9**
>
>(**Regla del sandwich**) Considérese tres sucesiones $\{a_n\}, \ \{b_n\}, \ \{c_n\}\subset\mathbb{R}$ verficando
>$$a_n\leq c_n\leq b_n, \quad \text{ para todo }n\in\mathbb{N}$$
>Si $\{a_n\}\to L$ y $\{b_n\}\to L$ para algún número real $L$, entonces $\{c_n\}\to L$.

>**Proposición 8.10**
>
>Se cumplen la siguientes propiedades:
>
>* Sean $\{a_n\}, \ \{b_n\}\subset\mathbb{R}$ sucesiones de números reales. Si $\{a_n\}$ es acotada y $\{b_n\}\to 0$, etonces
>$$\lim_{n\to\infty}a_nb_n=0$$
>
>*  Sean $\{a_n\}, \ \{b_n\}\subset\mathbb{R}$ sucesiones de números reales tales que $$|a_n|\leq b_n\quad\text{ para todo }n\in\mathbb{N}$$ Si $\{b_n\}\to 0$, entonces $\{a_n\}\to 0$.

## Funciones de una variable. Límites y continuidad

>**Teorema 9.1**
>
>Sea $f:D\longrightarrow\mathbb{R}$. Se tiene que
>
>* $f$ es **convexa** en $D$ si y sólo si su epígrafo es un conjunto convexo.
>* $f$ es **cóncava** en $D$ si y sólo si su hipógrafo es un conjunto convexo.

**Demostración (caso convexo)**

$(\Rightarrow)$

Supongamos que $f$ es convexa en $D$. Queremos ver que $\operatorname{epi}(f)$ es convexo.

Sean $(x_1, y_1), (x_2, y_2)\in\operatorname{epi}(f)$. Entonces
$$y_1\geq f(x_1), \quad y_2\geq f(x_2)$$
Tomemos $\lambda\in [0,1]$ y consideremos el punto intermedio:
$$(\bar x, \bar y)=(\lambda x_1 + (1-\lambda) x_2, \ \lambda y_1 + (1-\lambda)y_2)$$
Como $D$ es convexo (intervalo), se tiene
$$\bar x = \lambda x_1 + (1-\lambda) x_2\in D$$
Por convexidad de $f$,
$$f(\bar x) \leq \lambda f(x_1) + (1-\lambda) f(x_2)$$
Usando que $y_i\geq f(x_i)$,
$$\lambda f(x_1) + (1-\lambda) f(x_2)\leq \lambda y_1 + (1-\lambda)y_2=\bar y$$
Por tanto,
$$\bar y \geq f(\bar x)$$
y esto implica que $(\bar x, \bar y)\in \operatorname{epi}(f)$. Luego $\operatorname{epi}(f)$ es convexo.

$(\Leftarrow)$

Supongamos ahora que $\operatorname{epi}(f)$ es convexo. Probaremos que $f$ es convexa.

Sean $x_1, x_2\in D$. Entonces
$$(x_1, f(x_1)), \ (x_2, f(x_2)) \in \operatorname{epi}(f)$$
Por convexidad del epígrafo, para todo $\lambda\in [0,1]$ el punto
$$(\bar x, \bar y)=(\lambda x_1 + (1-\lambda)x_2, \lambda f(x_1)+(1-\lambda)f(x_2))$$
pertenece a $\operatorname{epi}(f)$.

Esto significa que 
$$\bar y \geq f(\bar x)$$
es decir, 
$$f(\lambda x_1 + (1-\lambda)x_2)\leq \lambda f(x_1)+(1-\lambda)f(x_2)$$
Esta es exactamente la definición de convexidad de $f$.

>**Teorema 9.2**
>
>Sea $f:D\longrightarrow\mathbb{R}$ una función convexa. Si $x_0\in D$ es un mínimo local de $f$, entonces dicho mínimo es, de hecho global.

**Demostración**

Sea $D\subset\mathbb{R}$ un conjunto convexo (en $\mathbb{R}$ esto equivale a que $D$ sea un intervalo) y sea $f:D\to\mathbb{R}$ convexa. Supongamos que $x_0\in D$ es un mínimo local de $f$. Entonces existe $\delta>0$ tal que
$$f(x_0)\le f(x)\qquad \text{para todo }x\in (x_0-\delta,x_0+\delta)\cap D$$

Tomemos un punto arbitrario $y\in D$. Si $y=x_0$, ya se tiene $f(x_0)\le f(y)$.

Supongamos ahora $y\ne x_0$ y consideremos, para $t\in[0,1]$, el punto del segmento entre $y$ y $x_0$:
$$x(t)=(1-t)y+t x_0$$

Como $D$ es convexo, $x(t)\in D$ para todo $t\in[0,1]$. Además,
$$x(t)-x_0=(1-t)(y-x_0)\quad\Rightarrow\quad |x(t)-x_0|=(1-t)|y-x_0|.$$

Sea
$$t_0=\max\left\{0,1-\frac{\delta}{|y-x_0|}\right\}\in[0,1)$$

Entonces, para todo $t\in(t_0,1]$ se cumple $(1-t)|y-x_0|<\delta$, y por tanto
$$|x(t)-x_0|<\delta\quad\Rightarrow\quad x(t)\in (x_0-\delta,x_0+\delta)\cap D$$

Por ser $x_0$ mínimo local, para todo $t\in(t_0,1]$ se tiene
$$f(x_0)\le f(x(t))$$

Por otra parte, como $f$ es convexa, para todo $t\in[0,1]$ se cumple
$$f(x(t))=f((1-t)y+t x_0)\le (1-t)f(y)+t f(x_0)$$

Uniendo ambas desigualdades, para $t\in(t_0,1)$ obtenemos
$$f(x_0)\le (1-t)f(y)+t f(x_0)$$

Restando $t f(x_0)$ a ambos lados,
$$(1-t)f(x_0)\le (1-t)f(y)$$

Como $t\in(t_0,1)$ implica $1-t>0$, podemos dividir por $1-t$ y concluir
$$f(x_0)\le f(y)$$

Dado que $y\in D$ era arbitrario, se deduce que
$$f(x_0)\le f(y)\quad\text{para todo }y\in D,$$
es decir, $x_0$ es un mínimo global de $f$.
